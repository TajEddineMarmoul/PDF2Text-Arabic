```python
import fitz
from pdf2text_arabic import extract_page, extract_pdf_result

doc = fitz.open("download/naw.pdf")
page = doc[57]  # printed page 58
```

```python
# on_empty="warn" (default) — page has image-based content → returns ""
# and logs a warning (because returning partial/missing content silently is unsafe)
print("=== on_empty='warn' (default) ===")
result = extract_page(page, on_empty="warn")
print(f"len={len(result)} repr={result!r}")
```

```python
# on_empty="ignore" — explicitly opt in to the text layer anyway
# (you acknowledge some image-based content is missing)
print("=== on_empty='ignore' ===")
print(extract_page(page, on_empty="ignore"))
```

```python
# on_empty="ocr" — mixed page: returns text layer + OCR'd image regions
# (table text gets recognized and spliced into the page output by y-position)
print("=== on_empty='ocr' ===")
print(extract_page(page, on_empty="ocr"))
```

```python
# Structured API — the way to check warnings programmatically
# (no need for logging handlers; inspect result.mixed_pages / result.warnings)
result = extract_pdf_result("download/naw.pdf", on_empty="warn")

print(f"pages_total    : {result.pages_total}")
print(f"pages_with_text: {result.pages_with_text}")
print(f"empty_pages    : {result.empty_pages}")
print(f"mixed_pages    : {result.mixed_pages}")
print(f"warnings       : {result.warnings}")
print()

# Act on warnings programmatically
if result.mixed_pages:
    print(
        f"⚠ {len(result.mixed_pages)} page(s) have image-based content: {result.mixed_pages}"
    )
else:
    print("✓ no mixed pages")

for w in result.warnings:
    kind, page_no = w.split(":")
    print(f"  warning kind={kind} page={page_no}")
```
