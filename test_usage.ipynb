# PDF2Text-Arabic — Usage Examples

Test notebook for the `pdf2text-arabic` package.


```python
# Create a small Arabic test PDF
import fitz

doc = fitz.open()
page = doc.new_page()

# Insert Arabic text
text_lines = [
    "بسم الله الرحمن الرحيم",
    "هذا نص تجريبي باللغة العربية",
    "يحتوي على عدة أسطر لاختبار الاستخراج",
    "رقم الصفحة 123 والسنة 2024",
]

y = 72
for line in text_lines:
    page.insert_text((400, y), line, fontsize=14, fontname="helv")
    y += 30

doc.save("test_arabic.pdf")
doc.close()
print("Created test_arabic.pdf")
```

```python
# Test: extract full PDF
from pdf2text_arabic import extract_pdf

text = extract_pdf("test_arabic.pdf")
print(text)
```

```python
# Test: extract single page
import fitz
from pdf2text_arabic import extract_page

doc = fitz.open("test_arabic.pdf")
page_text = extract_page(doc[0])
doc.close()
print(page_text)
```

```python
# Test: extract with cropping
text_cropped = extract_pdf("test_arabic.pdf", crop_top=50, crop_bottom=30)
print("With cropping:")
print(text_cropped)
```
