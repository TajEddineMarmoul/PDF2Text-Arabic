# Dahir 1.13.50: Page 7 Native Extraction

This PDF is selectable, so we are using the **pdf2text-arabic** library's `extract_page` function to target only Page 7 accurately without modifying the library source code.


```python
import os
import sys
import fitz
from pathlib import Path

# Ensure the local library is in the path
sys.path.append(str(Path(os.getcwd())))

from pdf2text_arabic._extract import extract_page

pdf_path = "download/- ظهير شريف رقم 1.15.73 بنشر الاتفاق في ميدان التعاون العسكري (1)-1748340749503.pdf"
doc = fitz.open(pdf_path)
page = doc[6]  # Page 7 (0-indexed)

print(f"Extracting Page 7 from {os.path.basename(pdf_path)}...\n")

# Use the existing library function designed for single pages
text = extract_page(page)
doc.close()

print("--- EXTRACTED RAG CONTENT ---")
print(text)
```
